# A Python Analysis of NVIDIA Stock Performance and Growth Trends

## 1. Problem Definition
This project aims to analyse NVIDIA’s stock performance using Python. The main goal is to understand the stock's price trends, returns, and volitility and to provide a better understanding for investors or finance students that are interested in evaluating stock performance. Besides that, in order to further evaluate the stock, NVIDIA is compared with Apple and S&P500 for a more accurate analysis. This analysis is mainly for beginner investors or finance students who are interested in understanding stock performances and risks.

## 2. Data Source

The dataset used in this project was obtained from Yahoo Finance using the yfinance Python library. The stock selected is NVIDIA (NVDA). The data are publicly available and suitable for educational analysis. The data were accessed on 21 April 2026.

## 3. Data Loading
In this section, required Python libraries are imported and NVIDIA stock data are downloaded from Yahoo Finance.

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
nvda = yf.download("NVDA", start="2020-01-01", end="2026-01-01", progress=False)
nvda.head()
```

## 4. Data Cleaning and Preparation

The dataset was checked for missing values and cleaned before analysis. Only the main variables required for the project were kept.

```python
nvda = nvda.dropna()
nvda = nvda[['Open', 'High', 'Low', 'Close', 'Volume']]
nvda.head()
```

## 5. Data Analysis

### 5.1 NVIDIA Closing Price Trend

The first step in the analysis is to observe the overall trend in NVIDIA’s closing stock price over time.

```python
plt.figure(figsize=(12,6))
plt.plot(nvda.index, nvda['Close'])
plt.title("NVIDIA Closing Price Trend")
plt.xlabel("Date")
plt.ylabel("Closing Price (USD)")
plt.show()
```

The chart shows the overall trend of NVIDIA’s closing stock price from 2020 to 2026. 
From the chart, it can be concluded that the stock price has increased over the years, indicating strong growth. 
This suggests increasing investor confidence, possibly driven by demand in the technology and AI sectors.

### 5.2 Daily Returns

To better understand the stock’s day-to-day performance, daily returns are calculated.

```python
nvda['Daily Return'] = nvda['Close'].pct_change()
nvda[['Close', 'Daily Return']].head()
```

```python
plt.figure(figsize=(12,6))
plt.plot(nvda.index, nvda['Daily Return'])
plt.title("NVIDIA Daily Returns")
plt.xlabel("Date")
plt.ylabel("Daily Return")
plt.show()
```

The daily return represents the percentage change in NVIDIA’s stock price from one day to the next. 
It can be seen that there are fluctuations which indicates that the stock experiences periods of both gains and losses. 
This then shows the volatility of the stock, which is an important measure of investment risk.

### 5.3 Volatility Analysis

Volatility is used to measure the riskiness of NVIDIA’s stock returns.

```python
volatility = nvda['Daily Return'].std()
print("NVIDIA Daily Volatility:", volatility)
```

Volatility measures the degree of variation in the stock’s returns. 
A higher volatility means greater uncertainty and risk. 
The calculated volatility suggests that while NVIDIA offers strong growth potential, it also involves higher risk compared to more stable stocks.

### 5.4 Performance Comparison with Apple and the S&P 500

To make the analysis more meaningful, NVIDIA is compared with Apple and the overall market represented by the S&P 500.

```python
aapl = yf.download("AAPL", start="2020-01-01", end="2026-01-01", progress=False)
spy = yf.download("SPY", start="2020-01-01", end="2026-01-01", progress=False)
```

```python
comparison = pd.DataFrame()

comparison['NVDA'] = nvda['Close'] / nvda['Close'].iloc[0] * 100
comparison['AAPL'] = aapl['Close'] / aapl['Close'].iloc[0] * 100
comparison['SPY'] = spy['Close'] / spy['Close'].iloc[0] * 100
```

```python
plt.figure(figsize=(12,6))
plt.plot(comparison.index, comparison['NVDA'], label='NVIDIA')
plt.plot(comparison.index, comparison['AAPL'], label='Apple')
plt.plot(comparison.index, comparison['SPY'], label='S&P 500')
plt.title("Stock Performance Comparison (Normalized)")
plt.xlabel("Date")
plt.ylabel("Index Value (Start = 100)")
plt.legend()
plt.show()
```

The comparison chart shows the relative performance of NVIDIA, Apple, and the overall market (S&P 500). 
As shown above, it can be observed that NVIDIA has significantly outperformed both Apple and the market overall in 2020 to 2026 given the fact that both is a major technology company. This outperformance may be attributed to strong investor demand driven by developments in artificial intelligence and high-performance computing. However, the earlier volatility analysis suggests that these higher returns are associated with an increased risk which then highlights the trade-off between return and risk, which is an important consideration for investors.


## 6. Key Findings
- NVIDIA demonstrated strong price growth over the years.
- The stock showed higher returns compared to Apple and the broader market (S&P500)
- Daily returns indicated noticeable volatility, suggesting higher risk.
- Moving trends and performance suggest strong investor confidence, particularly driven by the AI sector.

## 7. Conclusion

In conclusion, NVIDIA has shown exceptional growth and strong performance compared to other major stocks and the overall market. 
However, this growth comes with higher volatility, indicating greater investment risk. Overall, NVIDIA represents a high-growth but higher-risk investment option. This highlights the trade-off between risk and return, where higher potential returns are associated with higher volatility.
